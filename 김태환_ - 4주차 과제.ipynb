
# 문제3. 비선형함수
- 난이도 : 하
- 관련 개념 : 비선형함수, Activation function
- 코드를 통해 풀어주세요.
- 각 문제당 5점

ReLU, Sigmoid, tanh 함수를 직접 구현하여 아래의 값을 구하여라.

- 1) ReLU(2)
- 2) Sigmoid(2)
- 3) tanh(2)
- 4) tanh(-2)

```python
# 여기에 풀어주시면 됩니다
def ReLU(x):
  return max(0, x)
def Sigmoid(x):
  return 1 / ( 1 + np.e ** (-x))
def tanh(x):
  return 2 * Sigmoid(2*x) - 1

print(ReLU(2))
print(Sigmoid(2))
print(tanh(2))
print(tanh(-2))
```


# 문제4. 데이터의 크기에 따른 결과
- 난이도 : 하 
- 관련 개념 : 모델의 크기, 데이터의 크기

아래의 코드는 MLP를 통해 분류(classification)를 진행한 결과이다. 각 3개의 모델의 크기는 동일하지만 데이터의 수가 각각 100개, 1000개, 10000개이다. 그에 따라 점수도 다르다. 아래의 결과와 관련하여 MLP(딥러닝 모델)에서 데이터의 수가 가지는 의미, 관계 등에 대해 자유롭게 서술하시오.


```python
import warnings
warnings.filterwarnings(action='ignore') 

from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

X, y = make_classification(n_samples=100, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=1)
clf = MLPClassifier(hidden_layer_sizes=10, random_state=1).fit(X_train, y_train)
print(f'sample size = 100 일 때, 점수 {clf.score(X_test, y_test)}')

X, y = make_classification(n_samples=1000, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=1)
clf = MLPClassifier(hidden_layer_sizes=10, random_state=1).fit(X_train, y_train)
print(f'sample size = 1000 일 때, 점수 {clf.score(X_test, y_test)}')

X, y = make_classification(n_samples=10000, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=1)
clf = MLPClassifier(hidden_layer_sizes=10, random_state=1).fit(X_train, y_train)
print(f'sample size = 10000 일 때, 점수 {clf.score(X_test, y_test)}')
```

- 여기에 정답을 써주시면 됩니다. (더블클릭하세요)
데이터의 크기가 커지면, 모델의 수용력이 동일할지라도
Train Error와 Test Error가 동시에 줄어들어서
일반화 차이Generalization Gap가 감소한다

# 문제5. 출력층
- 난이도 : 하
- 관련 개념 : 출력층에서 소프트맥스(softmax) 사용
- 코드로 풀어주시기 바랍니다.
    - 소프트맥스(softmax)함수를 꼭 작성해야합니다.
- 정답만 맞으면 됩니다. 확률(10점), 최종 학점(10점)

심층이는 MLP를 통해 기말고사 성적을 예측하는 모델을 만들었다. 최종 결과는 총 5가지 A,B,C,D,E 학점이다. 본인의 점수를 모델에 넣고 최종적으로 나온 결과는 [10.0, 15.0, 8.5, 17.5, 2.8]라고 한다.

심층이는 이 값을 그대로 이용하지 않고 소프트맥스 함수를 통해 각 학점이 나올 확률과 최종적으로 받을 학점을 알아보고자 한다.

각 학점이 나올 확률과 이에 따라 최종적으로 A~E중 받는 학점을 구하라.


```python
# 확률은 소프트맥스(softmax)를 이용하여 구해주세요
def softmax(list_x):
  answer = []
  exp_sum = sum([ np.e ** x for x in list_x])
  for x in list_x :
    answer.append(( np.e ** x ) / exp_sum )
  return answer

target = ['A','B','C','D','E']
last_value =  [10.0, 15.0, 8.5, 17.5, 2.8]
result = softmax(last_value)
print( result )
print( target[ result.index(max(result)) ] )
```
