# 문제2. 확률

- 난이도 : 하
- 관련 개념 : 로지스틱 시그모이드 (Logistic Sigmoid)
- 풀이과정없이 답만 작성하면 됩니다. (소수 셋째자리까지)
    - 손으로 풀어도 되고 코딩해서 푸셔도 됩니다.
- 부분점수 있습니다. (각 문제마다 10점)

$x=3$에서 로지스틱 시그모이드 함수값과 미분했을 때의 값을 구하라.

1) $f(x)$

2) $f'(x)$

- 참고
    - 로지스틱 시그모이드 함수 $f(x) = \frac{1}{1+\exp(-x)}$

1) 0.9525741268224334 2) 0.045176659730912

```python
import numpy as np
```

```python
def sigmoid(x): 
  result = 1/(1+np.exp(-x)) 
  return result
```

```python
def derivative_sigmoid(x): 
  result = sigmoid(x) * (1 - sigmoid(x)) 
  return result
```

```python
sigmoid(3)
```

```python
derivative_sigmoid(3)
```

# 문제4. 수치계산

- 난이도 : 하
- 관련 개념 : overflow
- python 코드를 작성해주세요.

아래의 소프트맥스(softmax)함수를 실행시키면 overflow가 발생한다. 이를 보안하여 `upgrade_softmax` 함수를 작성하여라.

- 참고
    - 책에 해결방안이 나와있습니다.

```python
import numpy as np

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))

x = [1,2,3,4,1000]

softmax(x)
```

```python
# 해결방안

import numpy as np

def upgrade_softmax(x):
    c = np.max(x)
    exp_a = np.exp(x-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
x = [1,2,3,4,1000]

upgrade_softmax(x)
```
